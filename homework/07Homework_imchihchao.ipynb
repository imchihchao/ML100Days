### 1. 請比較對一個 1000 x 100 x 100 的陣列，使用不同方法對每一個元素 +1 的時間比較。




```python
# 記得先 Import 正確的套件
import numpy as np
```

```python
Z = np.random.randint(0, 10, 1000000).reshape(100, 100, 100)
```

```python
%timeit -n 10 a = 2
l = [[[k + 1 for k in j] for j in i] for i in Z]
```

```python
%timeit -n 10 a = 2

for i in Z.flat:
    i = i + 1
```

```python
%timeit -n 10 a = 2

for i in np.nditer(Z):
    i = i + 1
```

### 2. 如何從一個陣列中，找出出現頻率最高的數值與位置？


```python

Z = np.random.randint(0,10,50)
print(Z)
```

```python
mx = np.argmax(np.bincount(Z))
print(mx)
```

```python
print(np.where(Z == mx)[0])
```

### 3. 如何利用 list(...) 實現 a.tolist() 的效果？試著用程式實作。


```python
a = np.random.randint(10, size=6) 

print(a.tolist())
print(list(a))


b = np.random.randint(10, size=(3,4)) 

print(b.tolist())
print(list(b))


c = np.random.randint(10, size=(2,3,2)) 

print(c.tolist())
print(list(c))
```

```python
def tolist(na):
    return list(na) if na.ndim == 1 else [tolist(i) for i in na]

print(tolist(a))
print(tolist(b))
print(tolist(c))
```
