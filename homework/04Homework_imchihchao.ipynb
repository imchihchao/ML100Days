### 1. [簡答題] 請問下列程式碼，運算結果分別為何？

```

```


````python
import numpy as np

a = np.array( [20,30,40,50] )
b = np.array( [1,2,3,4] ) 
c = 1
d = np.array( [1] )
e = np.array( [1,2] )

print(a + a)
print(a + b)
print(a + c)
print(a + d)
print(a + e)
````

### 2. 如何在不用迴圈的情況下計算 (A+B)*(-A/2) ？那用迴圈怎麼做？


````python
# 記得先 Import 正確的套件

import numpy as np
````

````python
A = np.ones(3)*1
B = np.ones(3)*2

print((A+B)*(-A/2))

C = np.empty(3)
for i in range(3):
    C[i] = (A[i]+B[i])*(-A[i]/2)
print(C)
````

### 3. 請問如何計算「1x6 的單位矩陣」和「6x1 的單位矩陣」的內積和外積？


````python
A = np.ones((1,6))
B = np.ones((6,1))

print(A * B)
print(A @ B)
````
