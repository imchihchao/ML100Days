````python
# 記得先 Import 正確的套件
import numpy as np
````

### 3. 承上題，怎樣可以限制整數的範圍介於 m - n 之間？請將給定的 a 陣列當中的元素的範圍調整成 m - n 之間。


````python
m = 20
n = 40

a = np.random.rand(2, 3)
a = (a * (n - m) + m).round()

print(a)
````
