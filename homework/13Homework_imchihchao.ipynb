````python
# 記得先 Import 正確的套件
import numpy as np
import pandas as pd
````

### 1. 請接下列資料依照指定規則做合併：

```
- 依照 fruit 欄位做合併
- 依照 index 索引做合併
```


````python
df1 = pd.DataFrame({
    'fruit': ['apple', 'banana', 'orange'] * 3,
    'weight': ['high', 'medium', 'low'] * 3,
    'price': np.random.randint(0, 15, 9)
})

df2 = pd.DataFrame({
    'fruit': ['apple', 'orange', 'pine'] * 2,
    'weight': ['high', 'low'] * 3,
    'price': np.random.randint(0, 15, 6)
})
print(df1)
print(df2)
````

````python
# - 依照 fruit 欄位做合併

pd.merge(df1, df2, on='fruit')
````

````python
pd.merge(df1, df2)
````

````python
# - 依照 index 索引做合併
df1.join(df2)
````

### 2. [簡答題] 承上題，請問為什麼依照 merge 合併後有些資料不見了？


merge 預設的合併方式需要兩個 dataframe 的資料都存在才會保留進行合併，只存在任一邊的資料就不見了。

### 3. [簡答題] 承上題，請問為什麼依照 index 合併會發生錯誤？請用程式解決。

因為二個 dataframe 的欄位重覆所以無法進行合併。

````python
df1.join(df2, lsuffix='df1', rsuffix='_df2')
````
