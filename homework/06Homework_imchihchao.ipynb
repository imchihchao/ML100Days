### 1. [簡答題] 請問下列這三種方法有什麼不同？

```
print(a.sum()) 
print(np.sum(a))
print(sum(a)) 
```



np.sum(a) 及 sum(a) 是 numpy 中陣列的函數。  
sum(a) 則是 python 的內建函數。

### 2. 請對一個 5x5 的隨機矩陣作正規化的操作。


````python
# 記得先 Import 正確的套件
import numpy as np
````

````python
A = np.random.random((5, 5))
A = (A - A.min()) / (A.max() - A.min())
print(A)
````

### 3. 請建立一個長度等於 10 的正整數向量，並且將其中的最大值改成 -1。


````python
A = np.random.randint(0, 100, 10)
print(A)
i = A.argmax()
A[i] = -1
print(A)
````
