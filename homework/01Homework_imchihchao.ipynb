### 2. 請試著在程式中印出以下三個 NdArray 的屬性並且解釋結果？（屬性：ndim、shape、size、dtype、itemsize、length、type）

```
a = np.random.randint(10, size=6)
```

````python
# 記得先 Import 正確的套件
import numpy as np
````

````python

a = np.random.randint(10, size=6) 
print(a)
print(a.ndim)
print(a.shape)
print(a.size)
print(a.dtype)
print(a.itemsize)
print(len(a))
print(type(a))
'''
a的維度是1
a一維大小是6
a的大小是6個
a元素的資料型態是int64
a元素佔8個byte
a長度是6
a資料型態是numpy.ndarray
'''
````

````python
b = np.random.randint(10, size=(3,4)) 
print(b)
print(b.ndim)
print(b.shape)
print(b.size)
print(b.dtype)
print(b.itemsize)
print(len(b))
print(type(b))
'''
b的維度是2
b的維度大小是3*4
b的大小是12個
b元素的資料型態是int64
b元素佔8個byte
b長度是3
b資料型態是numpy.ndarray
'''
````

````python
c = np.random.randint(10, size=(2,3,2)) 
print(c)
print(c.ndim)
print(c.shape)
print(c.size)
print(c.dtype)
print(c.itemsize)
print(len(c))
print(type(c))
'''
c的維度是3
c的維度大小是2*3*2
c的大小是12個
c元素的資料型態是int64
c元素佔8個byte
c長度是2
c資料型態是numpy.ndarray
'''
````

### 3. 如何利用 list(...) 實現 a.tolist() 的效果？試著用程式實作。


````python
a = np.random.randint(10, size=6) 

print(a.tolist())
print(list(a))
````

````python
b = np.random.randint(10, size=(3,4)) 

print(b.tolist())
print([list(i) for i in b])
````

````python
c = np.random.randint(10, size=(2,3,2)) 

print(c.tolist())
print([[list(j) for j in i] for i in c])
````

````python
def tolist(na):
    if na.ndim == 1:
        return list(na)
    else:
        return [tolist(i) for i in na]

print(tolist(a))
print(tolist(b))
print(tolist(c))
````
